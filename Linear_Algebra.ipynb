Because we are going to take linear combinations.

We can also represent variables in a matrix.

#### Row Vector

In Row Vector there will be 1 row and multiple coloumns.

Example:

 \[
    C=
    \left[{\begin{array}{c}
    1 & 2 
    \end{array}}\right]
 \]
 
 Dimension = 1 * 3

#### Coloumn Vector

In coloumn vector there can be multiple rows but only one coloumn.

Example:

Coloumn Vector:
 \[
    B=
    \left[{\begin{array}{c}
    1\\
    3\\
    5\\
    \end{array}}\right]
 \]
 
 Dimension = 3 * 1

#### Linear Combinations

For Example: 
 We need to find linear combination of

\[
    X
    \left[{\begin{array}{c}
    2\\
    -1\\
    \end{array}}\right] + 
    Y
    \left[{\begin{array}{c}
    -1\\
    2\\
    \end{array}}\right] =
    \left[{\begin{array}{c}
    0\\
    3\\
    \end{array}}\right]
 \]

<br>
We can solve this in equation form or in matrix form 
<br>

By equation,

$$2X - Y = 0$$
$$-X + 2Y = 3$$

In this example there are few variables but in general there are many variables so we use matrix to solve becuase it will be fast.

 

#### Non Signular Matrix

If we are able to solve every X & Y for every output then the matrix is known as Non Singular Matrix.


#### Singular Matrix

If we are not able to solve every X & Y for every output then the matrix is known as Singular Matrix.

#### Operations we perform on Matrix

 #### 1. Elimination Matrix
 #### 2. Permutation Matrix

Let's just take an example we have 3 equationas and we need to find the values of X,Y,Z.

$$X + 2Y + Z = 2$$

$$3X + 8Y + Z = 12$$

$$4Y + Z = 2 $$
 
 <br>
 
 #### Sequence of Operations
 
 Step 1 
 
To solve these equation we can write these equation in a matrix form.

\[
    \left[{\begin{array}{c}
    1 & 2 & 1\\
    3 & 8 & 1\\
    0 & 4 & 1\\
    \end{array}}\right]\cdot
    \left[{\begin{array}{c}
    X\\
    Y\\
    Z\\
    \end{array}}\right] = 
    \left[{\begin{array}{c}
    2\\
    12\\
    2\\
    \end{array}}\right]
 \]
 
 
 
 

Step 2

Convert the matrix into Upper Triangular Matrix.

Make all the elements below the diagonal elements zero.

<br>
(a) After applying operation in second row($R_2$) $$R_2 = R_2 - 3R_1$$

\[
    \left[{\begin{array}{c}
    1 & 2 & 1\\
    0 & 8 & 1\\
    0 & 4 & 1\\
    \end{array}}\right]\cdot
    \left[{\begin{array}{c}
    X\\
    Y\\
    Z\\
    \end{array}}\right] = 
    \left[{\begin{array}{c}
    2\\
    6\\
    2\\
    \end{array}}\right]
 \]
 
 (b) Operation in $R_3$
 $$R_3 = R_3 - 2R_2$$
 
 \[
    \left[{\begin{array}{c}
    1 & 2 & 1\\
    0 & 2 & -2\\
    0 & 0 & 1\\
    \end{array}}\right]\cdot
    \left[{\begin{array}{c}
    X\\
    Y\\
    Z\\
    \end{array}}\right] = 
    \left[{\begin{array}{c}
    2\\
    6\\
    -2\\
    \end{array}}\right]
 \]

<br>
This method converting matrix into Upper Triangular matrix is known as Gaussian Elimination Method.
